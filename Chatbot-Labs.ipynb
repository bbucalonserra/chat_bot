### 2.1 Mapping inputs to responses using dictionaries.

We now use dictionaries to map user intents. Instead of writing a new *dictionary*, use the **ZIP** function on a list. 

```python
# Write a function that takes in precept categories and responses lists and returns a mapping between them.

# - Use ZIP function to combine two independent lists (precepts and responses) and create a new map between them.
# - Search on the ZIP function and how to use it in a for loop
# - Run the main loop and check that this function works

# ___ Write your code here _____

def zip_responses(precepts, response):
    mapping = {}
    for precept, response in zip(precepts, responses):
        mapping[precept] = response
    return mapping

result = zip_responses(precepts, responses)
print(result)
```

### 2.2 Using regular expressions to match patterns.

- Match complete words using regular expressions. 
- Example, match all occurrences of hello, hi, hey etc under a pattern called greetings.
- Use "|" symbol to string together multiple words.
- Each pattern should have atleast or 5 different occurrences of the word you want to match.
- Create atleast 5 different patterns.

---

```python
# Write a function to create 5 RE patterns and return them as separate variables.
# Give the function a characteristic name (e.g., re_patterns) 
# ___ Write your code here _____

def re_patterns():
    # 1. Cumprimentos
    pattern_greeting = re.compile(r"hello|hi|hey|greetings", re.IGNORECASE)

    # 2. Agradecimento
    pattern_thanks = re.compile(r"thank(s| you)?", re.IGNORECASE)

    # 3. Negação
    pattern_no = re.compile(r"\bno\b|\bnot\b|\bnope\b", re.IGNORECASE)

    # 4. Local (café)
    pattern_cafe = re.compile(r"cafe|coffee|tea", re.IGNORECASE)

    # 5. Despedida
    pattern_bye = re.compile(r"bye|goodbye|see you", re.IGNORECASE)

    return pattern_greeting, pattern_thanks, pattern_no, pattern_cafe, pattern_bye


# Teste
greet, thanks, no, cafe, bye = re_patterns()
print(greet.pattern)
print(thanks.pattern)
print(no.pattern)
print(cafe.pattern)
print(bye.pattern)
```

```python
## Detect pattern - Write a function to detect RE patterns in user input and return response category.

## Create the detect_pattern function with one argument : user_input
## 1. Call the re_patterns function and store patterns in seperate variables
## 2. Write if-else conditions to check for pattern match
##    2.1 For the if condition, use the re.search to check if the pattern exists in user input 
##    2.2 If there is a match, return the response type (e.g., "greetings" or "bye" or "happy")
##    2.3 Note: Use <span style="color:red; font-size: 14px"> elif </span> to match multiple conditions
##    2.4 If none of the conditions, return "default" which will be the default response of the chatbot
##    2.5 <span style="color:red; font-size: 14px"> Note: use else for last conditon </span>


# Give the function a characteristic name (e.g., detect_pattern) 
# ___ Write your code here _____

def detect_pattern(user_input):
    # Desempacotando as funções de re_patterns.
    pattern_greeting, pattern_thanks, pattern_no, pattern_cafe, pattern_bye = re_patterns()

    # 2. Testar cada padrão com re.search
    if re.search(pattern_greeting, user_input):
        return "greetings"
    elif re.search(pattern_thanks, user_input):
        return "thanks"
    elif re.search(pattern_no, user_input):
        return "no"
    elif re.search(pattern_cafe, user_input):
        return "cafe"
    elif re.search(pattern_bye, user_input):
        return "bye"
    else:
        # Caso nenhum padrão seja encontrado
        return "default"
    

# Teste
print(detect_pattern("Hello there!"))
print(detect_pattern("thank you so much"))
print(detect_pattern("no, not today"))
print(detect_pattern("Is there a cafe nearby?"))
print(detect_pattern("ok bye"))
print(detect_pattern("something random"))
```

```python
## Chatbot response - Write a function to return a responses by indexing dictionary with response categories.

## Create the chatbot_response function with one argument : user_input
## 1. Call the detect_pattern function and store the response in variable (e.g., detected_type)
## 2. Call the re_responses function and store responses in a variable (e.g., re_response)
## 2. Write if-else conditions to check for pattern match
##    2.1 For the if condition, check if the detected_type is one of the keys of the dictionary
##    2.2 If there is a match, return the response from dictonary
##    2.3 Note: Use elif to match multiple conditions
##    2.4 If none of the conditions, use "default" response

# Give the function a characteristic name (e.g., detect_pattern) 
# ___ Write your code here _____
def re_responses():
    # Dicionário que mapeia os tipos detectados (chaves) para respostas (valores)
    responses = {
        "greetings": "Hello! How can I assist you today?",
        "thanks": "You are welcome!",
        "no": "Is there anything else I can help you with?",
        "cafe": "There are 2 cafes - in main building and near the lecture halls.",
        "bye": "Goodbye! Have a great day!",
        "default": "Sorry, could you rephrase that again?"
    }
    return responses


def chatbot_response(user_input):
    # 1. Detectar o tipo de padrão presente no texto do usuário
    detected_type = detect_pattern(user_input)

    # 2. Obter o dicionário de respostas
    re_response = re_responses()

    # 3. Escolher a resposta apropriada com base no tipo detectado
    if detected_type in re_response:
        # se o tipo está no dicionário, retorna a resposta correspondente
        return re_response[detected_type]
    else:
        # caso nenhum padrão tenha sido reconhecido
        return re_response["default"]


# Teste
print(chatbot_response("hello there"))        # → Hello! How can I assist you today?
print(chatbot_response("thank you so much"))  # → You are welcome!
print(chatbot_response("no, not really"))     # → Is there anything else I can help you with?
print(chatbot_response("Is there a cafe?"))   # → There are 2 cafes...
print(chatbot_response("ok bye"))             # → Goodbye! Have a great day!
print(chatbot_response("random sentence"))    # → Sorry, could you rephrase that again?
```

```python
def chat():
    print("Chatbot: Hello! How can I help you today?")
    while True:
        user_input = input("You: ")

        # condição de saída
        if user_input.lower() in ["bye", "exit", "quit"]:
            print("Chatbot: Thank you for your interaction. Goodbye!")
            break

        # gera a resposta usando as funções anteriores
        bot_response = chatbot_response(user_input)
        print(f"Chatbot: {bot_response}")


# Start chatting with the bot
chat()
```

#### Writing advanced regular expressions

Let's exploring advanced REs by creating two more patterns to match the name and favourite color of the user. 

As name, and color can be any value, pattern matching needs to be generic and requires using alphabets [a-zA-Z], words [\w] and special symbols [*,+,?].

```python
# Write a regular expression pattern for a name and favourite color.

# 1. Names typically start with a letter in upper case followed by a sequence of letters in lower case.
# Try writing a few use cases to ensure that the pattern works.

# 2. Colors typically start with a pattern - color is <color_name> or I like <color_name>.
# Try writing a few use cases to ensure that the pattern works.




# Pattern for name
# Starts with uppercase letter followed by lowercase letters.
# Optionally allows a last name in the same format.
pattern_name = re.compile(r"^[A-Z][a-z]+(?: [A-Z][a-z]+)?$")

# Test name pattern
names = ["Bruno", "Maria Silva", "ana", "joão", "Joao da Silva"]
for n in names:
    print(n, "->", bool(re.match(pattern_name, n)))


# Pattern for favorite color
# Matches "color is <color>" or "I like <color>"
pattern_color = re.compile(r"(color is|I like) ([a-z]+(?: [a-z]+)?)", re.IGNORECASE)

# Test color pattern
colors = [
    "color is blue",
    "I like red",
    "Color is dark green",
    "I like light blue",
    "favorite color is purple"
]
for c in colors:
    match = re.search(pattern_color, c)
    if match:
        print(c, "-> color detected:", match.group(2))
    else:
        print(c, "-> no match")
```

### 3.1 Reading chatbot intents from a JSON file

Apply the file handling concepts and pandas to read JSON file (intents.json) and store its content.


```python
def read_json(filepath):
    with open(filepath, "r") as f:
        data = json.load(f)
    return pd.DataFrame(data["intents"])

df = read_json("intents.json")
df
```

```python
### Call the function and store intents in intents_json

## Write code here...
tags = df["tag"].tolist()
patterns = df["patterns"].tolist()
responses = df["responses"].tolist()


def store_patterns(tags, patterns):
    regex2int = dict()

    for tag, pattern_list in zip(tags, patterns): # O zip vai juntar dois elementos de uma lista em par, posição a posição.
        regex_pattern = r"\b(" + "|".join(pattern_list) + r")\b"
        regex2int[regex_pattern] = tag
    
    return regex2int

def store_responses(tags, responses):
    int2res = {}
    for tag, response_list in zip(tags, responses):
        int2res[tag] = response_list
    return int2res


regex2int = store_patterns(tags, patterns)
int2res = store_responses(tags, responses)

print(regex2int)
print(int2res)
```

```python
### Print print only the patterns of intents_json to see 
### their content as that's directly relevant for next step.
import re
import random

def generateResponse(user_input):
    for regex, intent in regex2int.items():
        if re.search(regex, user_input, re.IGNORECASE):
            return random.choice(int2res[intent])
    return "Sorry, could you rephrase that again?"

def chatbot():
    print("Chatbot ready! Type 'bye' to exit.")
    while True:
        user_input = input("You: ")
        if user_input.lower() == "bye":
            print("Chatbot: Goodbye!")
            break
        print("Chatbot:", generateResponse(user_input))

chatbot()
```
